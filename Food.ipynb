```python
import matplotlib.pyplot as plt
import numpy as np

x = range(7)

carbs = [3, 5, 5, 4, 3, 7, 6]

protein = np.flip([4, 5, 3, 2, 4, 5, 2], axis = 0)

fats = [1, 2, 2, 1, 3, 4, 4]
```

```python
plt.plot(x, carbs)
plt.plot(x, protein)
plt.plot(x, fats)
plt.legend(['carbs', 'protein', 'fats'])
plt.ylabel('Quantities')
plt.xlabel('Days of the week - Monday to Sunday')
```
